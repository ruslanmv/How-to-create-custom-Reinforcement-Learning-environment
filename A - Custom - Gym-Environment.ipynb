```python
import time
import matplotlib.pyplot as plt
import gym
```

```python
env = gym.make("worker:worker-v0")
```

```python
env.reset()
env.render("human")
action = env.action_space.sample()
img, reward, done, info = env.step(1)
print(reward, done, info)
plt.figure()
plt.imshow(img)
```

```python
env.reset()
for i in range(100):
    env.render("human")
    action = env.action_space.sample()
    img, reward, done, info = env.step(action)
    time.sleep(0.1)
```
